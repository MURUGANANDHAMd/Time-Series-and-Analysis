```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gold_data = pd.read_csv('/content/Gold Price.csv')
```

```python
gold_data.head()
```

```python
print(gold_data.columns)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(gold_data['Price'], label='Price')
plt.title('Price Trend')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
gold_data.iloc[0][['Open', 'High', 'Low', 'Price']].plot(kind='bar', figsize=(8, 5))
plt.title('Price Comparison for First Entry')
plt.ylabel('Value')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x=gold_data['Volume'], y=gold_data['Price'])
plt.title('Price vs Volume')
plt.xlabel('Volume')
plt.ylabel('Price')
plt.show()
```

```python
gold_data['Price'].hist(bins=20, edgecolor='black', figsize=(10, 6))
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
corr = gold_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
